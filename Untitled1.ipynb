# Predicting Metrics

```python
df = f_bpfip_
metric_inputs = ['FIP','_ab','wOBA']#,'wRAA+FIP','fWAR']
metric_plot = 'f_score'
lookback = 20
full_median = 40
lookback_metrics = ['f_score']
check_df = pd.DataFrame()
check_score = []
batters = df.unstack(level=1).index.tolist()
all_inputs = metric_inputs + lookback_metrics


for batter in batters:
    pass_cols = [metric_plot]+all_inputs if metric_plot not in all_inputs else all_inputs
    df_test = df.loc[batter][pass_cols].replace([np.inf, -np.inf], np.nan).dropna()
    df_test['game'] = np.arange(len(df_test))
    x_cols = []
    med_cols = []
    for met in lookback_metrics:
        for x in np.arange(1,full_median):
            column_name = '{}_+{}'.format(met,str(x))
            df_test[column_name] = df_test[met].shift(x)
            if x <= lookback:
                x_cols.append(column_name)
                med_cols.append(column_name)
            else:
                med_cols.append(column_name)
        column_name = '{}_mean'.format(met)
        df_test[column_name] = np.nanmean(df_test[med_cols].as_matrix(), axis=1)
        x_cols.append(column_name)
        column_name = '{}_std'.format(met)
        df_test[column_name] = np.nanstd(df_test[med_cols].as_matrix(), axis=1)
        x_cols.append(column_name)
    for met in metric_inputs:
        column_name = '{}'.format(met)
#         df_test[column_name] = df_test[met].shift(-1)
        x_cols.append(column_name)
    df_test['{}_-1'.format(metric_plot)] = df_test[metric_plot].shift(-1)
    df_test = df_test.dropna(subset=x_cols)
    if len(df_test) > lookback:
        X = df_test[['game'] + x_cols].as_matrix()
        y = df_test['{}'.format(metric_plot)].fillna(0)
        # Init regressor
        reg = LinearRegression(fit_intercept=True).fit(X, y)
#         reg = neural_network.MLPRegressor().fit(X, y)
        Y = reg.predict(X)
        # Add to check df and list for analysis
        check_df_sub = pd.DataFrame()
        check_df_sub[metric_plot] = df_test['{}'.format(metric_plot)]
        check_df_sub['p_{}'.format(metric_plot)] = reg.predict(X)
        check_df_sub['batter'] = batter
        check_df = check_df.append(check_df_sub)
        check_score.append(reg.score(X,y))
```

```python
z = np.polyfit(y, Y, 4)
p = np.poly1d(z)
xp = np.linspace(0, 70, 200)
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(check_df[metric_plot],check_df['p_{}'.format(metric_plot)], alpha=.15,s=5)
ax.plot([-10,70],[-10,70], color='r')
ax.plot(xp,p(xp), color='green', ls='--')
ax.set_xlim(-10,70)
ax.set_ylim(-10,70)
ax.set_xlabel(metric_plot)
ax.set_ylabel('p_{}'.format(metric_plot))
```

```python
df = f_bpfip_
metric_inputs = ['FIP']
metric_plot = 'f_score'
lookback = 5
full_median = 20
lookback_metrics = ['f_score']
check_df = pd.DataFrame()
check_score = []
batters = df.unstack(level=1).index.tolist()

pass_cols = [metric_plot] + metric_inputs if metric_plot not in metric_inputs else metric_inputs
df_test = df[pass_cols].replace([np.inf, -np.inf], np.nan).dropna()
x_cols = []
med_cols = []
for met in lookback_metrics:
    for x in np.arange(1,full_median):
        column_name = '{}_+{}'.format(met,str(x))
        df_test[column_name] = df_test[met].shift(x)
        if x <= lookback:
            x_cols.append(column_name)
            med_cols.append(column_name)
        else:
            med_cols.append(column_name)
    column_name = '{}_mean'.format(met,str(x))
    df_test[column_name] = np.nanmean(df_test[med_cols].as_matrix(), axis=1)
    x_cols.append(column_name)
for met in metric_inputs:
    column_name = '{}_-1'.format(met)
    df_test[column_name] = df_test[met].shift(-1)
    x_cols.append(column_name)
final = pd.DataFrame()
for batter in batters:
    player_break = df_test.loc[batter]
    player_break_ = player_break.tail(len(player_break)-5)
    final = final.append(player_break_)
print(len(final))
final = final.dropna(subset=x_cols)
X = final[x_cols].as_matrix()
y = final['{}'.format(metric_plot)].fillna(0)
# Init regressor
reg = LinearRegression(fit_intercept=True).fit(X, y)
Y = reg.predict(X)
# Add to check df and list for analysis
check_df_sub = pd.DataFrame()
check_df_sub[metric_plot] = final['{}'.format(metric_plot)]
check_df_sub['p_{}'.format(metric_plot)] = reg.predict(X)
check_df_sub['batter'] = batter
check_df = check_df.append(check_df_sub)
check_score.append(reg.score(X,y))
```
